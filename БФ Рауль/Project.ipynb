#### Возраст кандидатов

Посчитаем количество кандидатов разных возрастов. Для этого создадим категории возрастов:
- до 14 лет;
- от 14 до 18;
- от 18 до 23;
- от 23 до 35;
- от 35.

```python
# создадим критерии для создания категорий
conditions = [
    (df['возраст'] < 14),
    ((df['возраст'] >= 14) & (df['возраст'] <= 17)),
    ((df['возраст'] >= 18) & (df['возраст'] <= 23)),
    ((df['возраст'] >= 24) & (df['возраст'] <= 35)),
    (df['возраст'] > 35)
]


# создадим названия категорий
categories = ['<14', '14-17', '18-23', '24-35', '>35']

# создадим столбец с названием категории возраста
df['age_category'] = np.select(conditions, categories)

# посчитаем количество кандидатов в каждой категории
category_count = (
    df.groupby('age_category', as_index=False)
    .agg({'id_кандидата': 'nunique'})
    .rename(columns={'id_кандидата': 'количество_кандидатов'})
    .sort_values(by='количество_кандидатов', ascending=False)
)

print(category_count)

# визуализируем результат
category_count.plot.pie(y='количество_кандидатов',
                        labels=category_count['age_category'],
                        ylabel='',
                        autopct='%.1f%%',
                        legend=False)

plt.title('Возраст кандидатов');
```

#### Анализ временных промежутков

Проведем анализ временных промежутков для различных событий и определим распределение кандидатов по категориям промежутков.

Будем исследовать:
1. Количество кандидатов по времени нахождения **первой** работы после добавления;
1. Количество трудоустройств по продолжительности работы до **завершения**;
1. Количество **работающих** на текущую дату по продолжительности работы;
1. Количество трудоустройств по продолжительности работы кандидатов.

Использовать будем следующие временные категории:
1. меньше суток;
1. до месяца;
1. до полугода;
1. более полугода.

Создадим функцию для подсчета разницы между датами в днях и месяцах.

```python
def date_diff(df, data_start, data_end, title):
    '''
    Данная функция принимает на входе датафрейм, дату начала и конца события, название графика.
    Счиает разницу между датами в днях и месяцах.
    Создает категории разницы дат.
    Возвращает количество кандидатов в категориях и визуализирует результат
    '''

    # найдем разницу между датами в днях
    df['days_diff'] = (df[data_end] - df[data_start]).dt.days

    # найдем разницу между датами в месяцах
    df['months_diff'] = (df[data_end] - df[data_start]) // np.timedelta64(1, 'M')

    # создадим критерии для создания категорий
    conditions = [
        ((df['days_diff'] >= 1) & (df['days_diff'] <= 30)),
        ((df['months_diff'] >= 1) & (df['months_diff'] < 3)),
        ((df['months_diff'] >= 3) & (df['months_diff'] < 6)),
        (df['months_diff'] >= 6)
    ]


    # создадим названия категорий
    categories = ['до месяца', 'от 1 до 3 месяцев', 'от 3 месяцев до полугода', 'более полугода']

    # создадим столбец с названием категории длительности работы
    df['category_job_duration'] = np.select(conditions, categories, default='меньше суток')

    # посчитаем количество строк в каждой категории
    category_count = (
        df['category_job_duration']
        .value_counts()
        .to_frame()
        .reset_index()
    )

    # визуализируем результат
    category_count.plot.pie(y='category_job_duration',
                            labels=category_count['index'],
                            ylabel='',
                            autopct='%.0f%%', 
                            legend=False)
    plt.title(title)

    return category_count
```

##### Количество кандидатов по времени нахождения первой работы после добавления

Найдем количество кандидатов по временным интервалам, за которые кандидаты находят свое **первое** трудоустройство после добавления.
